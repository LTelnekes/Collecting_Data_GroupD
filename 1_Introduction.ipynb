### **Part 1: Introduction**
Our idea for this project arose when the topic of 'Spotify Wrapped' came up in one of our classes, leading us to explore how Spotify and other popular music streaming platforms over the years have challenged traditional or mainstream popularity metrics within the music industry. While brainstorming research questions, we agreed that it would be beneficial to collect data from similar sources: in this case, both Spotify charts and the Billboard Hot 100 display “top songs," but the platforms are different enough to isolate any distinctions that might exist between the two.

In 2018, The Washington Post also noted that a shift in the charts had occured, with Billboard being producing more traditonal charts as opposed to how streaming platforms look at musical success. Whereas “Billboard’s charts used to be the barometer for music success,” the Washington Post questioned if this had become “meaningless in the streaming age?” ([Andrews, 2018](https://www.washingtonpost.com/news/arts-and-entertainment/wp/2018/07/05/billboards-charts-used-to-be-our-barometer-for-music-success-are-they-meaningless-in-the-streaming-age/)). The article notes that “since their inception in 1958, the Billboard charts served a window to pop music history. Along with statistics collected by RIAA and Nielsen, they offer a road map of what tunes, musicians and genres Americans found interesting enough to consume en masse” even in modern times ([Andrews, 2018](https://www.washingtonpost.com/news/arts-and-entertainment/wp/2018/07/05/billboards-charts-used-to-be-our-barometer-for-music-success-are-they-meaningless-in-the-streaming-age/)). Trying to “stay in front of the game” Billboard constantly reconsiders how they should respond to up and coming technologies and in 2012 they already added streaming as a metric in their chart. To summarize, “the company is always considering what a song download is worth, what the difference is between a stream and a radio play, etc.” ([Andrews, 2018](https://www.washingtonpost.com/news/arts-and-entertainment/wp/2018/07/05/billboards-charts-used-to-be-our-barometer-for-music-success-are-they-meaningless-in-the-streaming-age/)).

While one could say that both Spotify and Billboard charts display popular music, Spotify’s chart formulation is algorithmic, concerning mostly streaming but also shares, number of listeners, including new ones, and even rates of saving songs to playlists or skipping ([Charts, n.d.](https://support.spotify.com/us/artists/article/charts/)), while the Billboard Hot 100 charts are informed by several factors, including sales (physical and digital), radio airplay, digital downloads, and online streaming ([How Do Billboard Charts Work?, 2023](https://inside-the-industry.com/2023/04/21/how-do-billboard-charts-work-the-hot-100-explained/)). Spotify’s chart formulation is much more modern and based on displayed streams in this sense, while the Billboard Hot 100 uses more traditional processes and metrics. According to Billboard’s senior vice president of charts and data development Silvio Pietroluongo, “streaming changed the actual manner in which we listen to music . . . When streaming started, the idea was people would pick the tracks they wanted to hear, but now they’re being fed songs like a jukebox,’” leading to a constant reevaluation of how Billboard should be scoring song popularity to reflect the consumer's behavior in an accurate way ([Andrews, 2018](https://www.washingtonpost.com/news/arts-and-entertainment/wp/2018/07/05/billboards-charts-used-to-be-our-barometer-for-music-success-are-they-meaningless-in-the-streaming-age/)). This shows that Billboard has responded to the changes that occurred in the streaming age, but not without careful consideration. 

#### Goal of the Project

Music charts may hold valuable information that lies beyond what is first recognizable. The overall popularity of songs, artists, and genres may even reflect societal occurrences and their subsequent effects on a population, which can be seen in the study "Unemployment Rate Predicts Anger in Popular Music Lyrics: Evidence From Top 10 Songs in the United States and Germany From 1980 to 2017" by [Qiu, et al (2021)](https://doi.org/10.1037/ppm0000282). The arts, including music, hold important information about the psychological changes that occur alongside societal occurrences. While there has been much documentation about how historical occurrences have given way to the rise of popular genres and artists (i.e. social injustice to hip hop, post-war conservatism to Elvis' dance style), it is still crucial to examine consumer behavior, as it can lead to even more revelations.

Our project aims to highlight differences between two sources of tracking music popularity using different metrics of consumer behavior, providing a useful roadmap for a number of research projects for which this data collection process may be useful. Comparing traditional and modern music popularity metrics may bolster the data's usefulness, as there are significant differences in the demographics of consumers measured by Spotify and the Billboard Hot 100.

#### Methodology
So, how does this streaming and the platforms that emerged with it, lead to differences in music consumption? Both organizations use different barometers to account for their weekly charts, and in light of this project, this led us to ask the research question: <font color=#a467aa> **How does music popularity differ across the United States’ Spotify Charts and the Billboard Hot 100?** </font> As the Billboard Hot 100 is the music industry’s standard record chart in the United States, we decided to collect the Spotify charts data, focused on the United States as well, instead of globally. This then makes for a more equal comparative analysis than if we were to compare the Billboard charts to a global Spotify chart, enabling us to compare the platforms, instead of the difference between U.S. listening behavior vs that of the ‘world’. There could be many reasons why differences might exist, including the differing metrics just mentioned, the average ages of those who stream music over a digital platform like Spotify versus those who frequently listen to music over the radio or purchase it, or which charts/platforms artists choose to promote, to name a few factors. This research question opens the door for illuminating data collection. We have collected data from Spotify’s Weekly Top 200 charts throughout 2021 focused on the region ‘United States’ as well as the Billboard Hot 100 charts in 2021.
