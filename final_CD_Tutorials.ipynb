1) Download the weekly charts

To compile a comprehensive dataset for our analysis, we downloaded top 200 weekly charts spanning all 52 weeks of 2021 from the Spotify Charts website (https://charts.spotify.com/charts/overview/us), which archives weekly charts from various countries.

2) Combine weekly chart data and enrich the dataset

```python
import os
import pandas as pd
```

3) Define the columns we need for the weekly dataset

```python
col_weekly = ['rank', 'uri', 'artist_names','track_name',
              'peak_rank', 'previous_rank', 'weeks_on_chart', 'streams', 'date',
              'country']
```

4) Function to Combine Data from Multiple CSV Files, Add New Columns, and Extract Top 100 Songs

```python
def weekly_concat(directory):
    # Initialize an empty DataFrame with the specified columns
    df = pd.DataFrame(columns=col_weekly)

    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)

        # checking if it is a file
        if os.path.isfile(f):
            # Read the first 100 rows from each CSV file
            df_week = pd.read_csv(f, nrows=100)
            # Extract the date from the filename
            date = f[-14:-4]
            # Add 'date' and 'country' columns to the DataFrame
            df_week['date'] = date
            df_week['country'] = country
            df = pd.concat([df, df_week], ignore_index=True) 

    df.to_csv('Spotify_2021_weekly.csv', index=False) 

    return None
```

5) Specify the directory containing CSV files and set the country

```python
directory = 'Spotify_2021_weekly_charts'
country = 'US'
```

```python
# Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for nicer plots and make the colours more readable for colorblind people that may read the tutorial
sns.set(style="whitegrid")
sns.set_palette("colorblind", n_colors=10) 
```

#### Preprocess data

```python
# Convert 'date' column to datetime
spotify_df['date'] = pd.to_datetime(spotify_df['date'])
billboard_df['date'] = pd.to_datetime(billboard_df['date'])

# Aggregate for dataset
spotify_artist_stats = spotify_df.groupby('artist').agg({'streams': 'sum', 'rank': 'mean'})
billboard_artist_stats = billboard_df.groupby('artist').agg({'weeks_on_charts': 'count', 'rank': 'mean'})

# Merge the datasets
combined_df = pd.merge(spotify_artist_stats, billboard_artist_stats, 
                       left_index=True, right_index=True)  # Adjust left_on and right_on if necessary

# Rename columns for clarity
combined_df.rename(columns={'rank_x': 'Spotify_Avg_Rank', 'rank_y': 'Billboard_Avg_Rank',
                            'streams': 'Spotify_Total_Streams', 
                            'weeks_on_charts': 'Billboard_Total_Weeks'}, inplace=True)
```

<div class="alert-info">
    <h3><strong>A) Create a Scatter Plot to Compare Average Ranks on Spotify and Billboard</strong></h3>
</div>


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
# Assuming 'Total_Streams' is a column in your dataframe that represents the total number of streams for a song
point_sizes = combined_df['Spotify_Total_Streams'] / 1000  # Scale down for better visualization
sns.scatterplot(data=combined_df, x='Spotify_Avg_Rank', y='Billboard_Avg_Rank', size=point_sizes, legend=False, alpha=0.7)
plt.title('Average Rank Comparison: Spotify vs Billboard')
plt.xlabel('Spotify Average Rank')
plt.ylabel('Billboard Average Rank')
plt.gca().invert_xaxis() # Inverts the X-axis for better visualization

sns.regplot(data=combined_df, x='Spotify_Avg_Rank', y='Billboard_Avg_Rank', scatter=False, color='red')
plt.show()
```

In the scatter plot, each point represents an artist. The position of the point is determined by the artist's average rank on Spotify and Billboard. Points closer to the bottom right corner indicate artists who have a high average rank (lower numerical value, which is better) on both platforms. From the scatter plot, we can observe that there is a spread of points across the plot, but there doesn't seem to be a clear, strong correlation between Spotify and Billboard rankings. While there are a few points towards the bottom right, suggesting some artists rank highly on both platforms, the majority of the points are scattered, indicating variability in how artists are ranked between Spotify and Billboard.
```python
# Import Libraries
import matplotlib.pyplot as plt

# Create a list of unique artists from both charts
all_artists = list(set(list(top_artists_spotify.keys()) + list(top_artists_billboard.keys())))

# Initialize variables for plotting
positions = range(len(all_artists))
bottom_spotify = [0] * len(all_artists)
bottom_billboard = [0] * len(all_artists)

# Create a figure and axis
plt.figure(figsize=(12, 6))

# Plot Spotify data
plt.bar(positions, [top_artists_spotify.get(artist, 0) for artist in all_artists],
        color='#298c8c', label='Spotify', alpha=0.6)

# Plot Billboard data on top of Spotify data
plt.bar(positions, [top_artists_billboard.get(artist, 0) for artist in all_artists],
        bottom=bottom_spotify, color='#800074', label='Billboard', alpha=0.6)

# Update bottom values for the next set of bars
bottom_spotify = [bottom_spotify[i] + top_artists_spotify.get(artist, 0) for i, artist in enumerate(all_artists)]
bottom_billboard = [bottom_billboard[i] + top_artists_billboard.get(artist, 0) for i, artist in enumerate(all_artists)]

# Set labels and legend
plt.xlabel('Artists')
plt.ylabel('Number of Songs')
plt.title('Top Performing Artists')
plt.xticks(positions, all_artists, rotation='vertical')
plt.legend()

# Show the plot
plt.show()
```

    <h3><strong>E) Create a Heatmap of Artist Presence on Spotify and Billboard Charts</strong></h3>

```python
# Rename rank columns for clarity
spotify_df.rename(columns={'rank': 'rank_spotify'}, inplace=True)
billboard_df.rename(columns={'rank': 'rank_billboard'}, inplace=True)

# Print column names to verify the changes
print("Spotify DataFrame columns:", spotify_df.columns)
print("Billboard DataFrame columns:", billboard_df.columns)

# Merge datasets on artist and song
combined_df = pd.merge(spotify_df, billboard_df, on=['artist', 'song'], how='outer')

# Print column names of the combined DataFrame
print("Combined DataFrame columns:", combined_df.columns)
```

```python
# Filter out rows where either 'rank_spotify' or 'rank_billboard' is NaN, ensure we only analyze artists who have rankings on both platforms
filtered_df = combined_df.dropna(subset=['rank_spotify', 'rank_billboard'])

# Create a pivot table with the filtered data
artist_pivot = filtered_df.pivot_table(index='artist', 
                                       values=['rank_spotify', 'rank_billboard'], 
                                       aggfunc='mean')

# Fill any NaN values, if necessary (should be minimal or none after filtering)
artist_pivot.fillna(0, inplace=True)

# Plot the heatmap with an adjusted figure size
plt.figure(figsize=(10, 20))  # Adjust the size as needed based on the number of artists
sns.heatmap(artist_pivot, annot=True, cmap='viridis')
plt.title('Heatmap of Artist Presence on Both Spotify and Billboard Charts')
plt.xlabel('Chart')
plt.ylabel('Artist')
plt.show()
```

The overall distribution of colors across the heatmap gives a visual representation of how artists compare overall on the two platforms. The intensity of the color correlates with the rank's value. Darker purple shades indicate a higher position (closer to 1), and brighter green/yellow ones suggest a lower position (further away from 1). By comparing the color intensities for each artist across both platforms, we can determine where an artist is more popular or has a better ranking. For instance, some artists like Camila Cabello Justin Bieber, Masked Wolf,  have relatively balanced rankings, but some artists, for instance,  Andy Williams and Ariana Grande have a significantly higher ranking on Spotify compared to Billboard.  In addition, Adele has an average rank of 23 on Spotify and 6.4 on Billboard, indicating a better average position on Billboard charts. This visualization can help us undestand whose musice  might be more popular or better promoted on Billboard compared to Spotify (or the other way round).
